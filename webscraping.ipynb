```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import re
pages=["https://getir.com/kategori/temel-gida-IQH9bir3bX/pirinc-YHe3qmOWZ4/"]
for page in pages:
    options = Options()
    options.headless = False

    # WebDriver'ı başlat
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)

    # URL'ye git
    url = page
    driver.get(url)


    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    price_elements = soup.find_all('span', class_='style__Text-sc-__sc-1nwjacj-0 grTSzD sc-c016d6c1-6 jmELbE')


    prices = []
    for element in price_elements:
        price_text = element.text
        price_str = re.findall(r'₺([\d,]+)', price_text)[0]
        price_float = float(price_str.replace(',', '.'))
        pirincfiyatlarım.append(price_float)

    product_name_elements = soup.find_all('span', class_='style__Text-sc-__sc-1nwjacj-0 gXitDI sc-c016d6c1-4 jFvYQy')

    # Ürün adlarını çek
    product_names = [element.text for element in product_name_elements]
    pirinç_names.extend(product_names)

    driver.quit()

```
