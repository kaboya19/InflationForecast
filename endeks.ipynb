```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("harcama gruplarina gore endeks sonuclari.xlsx")
data
```

```python
data.columns=data.iloc[2,:].values
```

```python
data=data.iloc[4:,:]
data
```

```python
temmuz=100*((data.iloc[-1,19:]/data.iloc[-2,19:])-1)
temmuz=temmuz.sort_values()
temmuz=temmuz.drop_duplicates()
temmuz
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(12,8))
plt.barh(y=temmuz.tail(20).index,width=temmuz.tail(20).values)
```

```python
sa=pd.read_excel("sa.xlsx")
sa=sa.dropna()
sa.columns=["Tarih","MA TÜFE"]
sa=sa.set_index(pd.date_range(start="2006-01-31",periods=len(sa),freq="M"))
sa=sa.drop("Tarih",axis=1)
sa["Aylık Arındırılmış Enflasyon"]=sa["MA TÜFE"].pct_change()*100
sa
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(12,8))
plt.title("Mevsimsellikten Arındırılmış Aylık Enflasyon,3 Aylık Hareketli Ortalama")
plt.plot(sa["Aylık Arındırılmış Enflasyon"].rolling(3).mean().tail(24))
plt.xticks(sa.rolling(3).mean().tail(24).index[::3],rotation=45)
```
